```python
import numpy as np
import matplotlib.pyplot as plt
import os
import PIL
PIL.Image.MAX_IMAGE_PIXELS = None
```

```python
ai_dir = r'ims\aerial_photos\RGB'
im_down = list()
for id in os.listdir(ai_dir)[90:96]:
    im = plt.imread(ai_dir + r'\\' + id)
    im_down.append(im[::20, ::20])

```

```python
save_dir = '../figures/quickviews'
for i, im in enumerate(im_down):
    fig, ax = plt.subplots()
    ax.imshow(im)
    ax.axis('off')
    fig.savefig(save_dir + '/qw_{}.png'.format(i), bbox_inches='tight', pad_inches=0.05)
```
