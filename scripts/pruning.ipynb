```python
import os
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
del_ids = []
for id in os.listdir(r'test_output/annot'):
    arr = np.load('test_output/annot/'+id, allow_pickle=True).item()
    id = id[:-4]
    if len(arr.keys()) > 10:
        del_ids.append(id)
print(len(del_ids))    
        
```

```python
filenames = os.listdir('test_output/annot')
for i, file in enumerate(filenames):
    filenames[i] = file[:-4]
for file in filenames:
    print(file)
```

```python
n_samples = len(filenames)
n_valid = int(0.2*n_samples)
random.shuffle(filenames)
valid_list = filenames[:n_valid] 
train_list = filenames[n_valid:]
```

```python
with open('test_output/all_list.txt', 'w') as file:
    for name in filenames:
        file.write(name+ '\n')

with open('test_output/valid_list.txt', 'w') as file:
    for name in valid_list:
        file.write(name+'\n')

with open('test_output/train_list.txt', 'w') as file:
    for name in train_list:
        file.write(name+ '\n')
```
